```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import random
import math
np.random.seed(0) 
```

```python
rows = []
with open('data.txt','r') as file:
    for line in file:
        line = np.array(line.split(' '))
        line = line.astype(np.float)
        rows.append(line)
rows = np.array(rows)
```

```python
data = StandardScaler().fit_transform(rows)
```

```python
def get_CoVariance(rows):    
    N = len(rows)
    mean = np.mean(rows, axis=0)
    rows_ = rows-mean
    S = np.dot(rows_.T, rows_.conj()) / (N-1)
#     print(S)
    eig_vals, eig_vecs = np.linalg.eig(S)
#     print("Eigen Values \n\n",eig_vals)
#     print("Eigen Vectors \n\n",eig_vecs)

    # Make a list of (eigenvalue, eigenvector) tuples
    eig_pairs = []
    for i in range(len(eig_vals)):
        eig_vals_abs = np.abs(eig_vals[i])
        eig_pairs.append((eig_vals_abs, eig_vecs[:,i]))

    eig_pairs.sort(key=lambda x: x[0], reverse=True)
    
#     print(eig_pairs[0][1])
    
    print(len(eig_pairs))
    matrix = np.zeros((len(eig_pairs), 2))
    
    for i in range(len(eig_pairs)):
        matrix[i][0] = eig_pairs[0][1][i]
        matrix[i][1] = eig_pairs[1][1][i]

    processed_data = np.dot(rows,matrix)

    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.scatter(processed_data[:,0], processed_data[:,1], facecolors='none', edgecolors = 'blue')
    return processed_data  

# get_CoVariance(data)
```

```python
def N_k(x_i, mean_k, covariance_k):
    xi_mean_k = x_i - mean_k
    xi_mean_k_T = np.transpose(xi_mean_k)
    exponent_part = -(1/2)* np.dot(np.dot(xi_mean_k_T , np.linalg.inv(covariance_k)), xi_mean_k)
    exponent_part = math.exp(exponent_part)
    root_part = math.pow(2*math.pi, D) * np.linalg.det(covariance_k)
    root_part = 1/ math.pow(root_part, 1/2)    
    return root_part*exponent_part
```

```python
def calculate_Loglikelihood(mean, co_variances, mixing_coefficients, processed_data, K):
    Loglikelihood = 0
    for i in range(len(processed_data)):
        sum_k = 0
        for k in range(K):
            sum_k += mixing_coefficients[k] * N_k(processed_data[i], mean[k], co_variances[k])
        ln = math.log(sum_k)
        Loglikelihood += ln
    return Loglikelihood
```

```python
def E_step(processed_data, mean,co_variances,mixing_coefficients, k):
    probabilities = [[0.0 for k_ in range(k)] for i in range(len(processed_data))]    
    for i in range(len(processed_data)):
        denominator = 0.0
        for k_ in range(k):
            probabilities[i][k_] = mixing_coefficients[k_] * N_k(processed_data[i], mean[k_], co_variances[k_])
            denominator += probabilities[i][k_]
        probabilities[i] = [prob/denominator for prob in probabilities[i]]
#     print(probabilities)
    return probabilities
```

```python
def M_step(probabilities, processed_data, k):
    mean = np.zeros((k,D))
    Sum_pk = np.ndarray.sum(np.array(probabilities), axis = 0)
#     print(len(Sum_pk))
#     print(Sum_pk)
#     # -----------------MEAN--------------------------
    for k_ in range(k):
        for i in range(len(processed_data)):
            mean[k_] += probabilities[i][k_] * processed_data[i]
        mean[k_] = [x/Sum_pk[k_] for x in mean[k_]]
    
#     print(mean)
    
    # -----------------co_variances--------------------------
    
    co_variances = [[[0 for i in range(D)] for i in range(D)] for k_ in range(k)]
#     print(co_variances)
    for k_ in range(k):
        for i in range(len(processed_data)):
            xi_meank = processed_data[i] - mean[k_]
            nom_temp = probabilities[i][k_]* np.outer(xi_meank, xi_meank.T) 
            co_variances[k_] += nom_temp
        co_variances[k_] /= Sum_pk[k_]
#     print(co_variances)
    
    # -----------------mixing_coefficients--------------------------
   
    mixing_coefficients = [x/len(processed_data) for x in Sum_pk]
    
#     print(mixing_coefficients)
    
    return mean, co_variances, mixing_coefficients
```

```python
def identity(n):
    m=[[0 for x in range(n)] for y in range(n)]
    for i in range(0,n):
        m[i][i] = 1
    return m
```

```python
def init(processed_data,k):
    mean = np.random.permutation(processed_data)[:k]    
    co_variances = []    
    for i in range(k):
        co_variances.append([i for i in identity(D)])
    mixing_coefficients = [1/k for i in range(k)]  
    return mean, co_variances, mixing_coefficients
# init(processed_data,5)
```

```python
def EM(processed_data, k=3):
    mean, co_variances, mixing_coefficients = init(processed_data,k)
    Loglikelihood = calculate_Loglikelihood(mean, co_variances, mixing_coefficients, processed_data, k)
    print("\nInitial Loglikelihood: ", Loglikelihood)
    Threshold = 0.0000001
    maxIteration = 500
    loop = 1
    while True:
        prevLog = Loglikelihood
        probabilities = E_step(processed_data, mean,co_variances,mixing_coefficients, k)
        mean, co_variances, mixing_coefficients = M_step(probabilities, processed_data, k)  
        Loglikelihood = calculate_Loglikelihood(mean, co_variances, mixing_coefficients, processed_data, k)
        print("For Loop :", loop, "Loglikelihood is :", Loglikelihood)
        loop = loop + 1
        if abs(Loglikelihood-prevLog)< Threshold or loop >= maxIteration:
            return mean, co_variances, mixing_coefficients 
```

```python
def PCA_EM_Clustering(processed_data,mean,co_variances, k=3):
    x=[]
    y=[]
    for i in range(k):
        x.append([])
        y.append([])
        
    for i in range(len(processed_data)):
        probabilities = np.zeros(3)
        for k_ in range(k):
            probabilities[k_] = N_k(processed_data[i], mean[k_], co_variances[k_])
        
        index = np.argmax(probabilities)
#         print(index)
        
        for k_ in range(k):
            x[index].append(processed_data[i][0])
            y[index].append(processed_data[i][1])
    
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    for k_ in range(k):
        plt.scatter(x[k_], y[k_])
```
